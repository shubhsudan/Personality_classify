```python
#Now We encode the personality labels
# Define the personality type mapping (from 16P-Mapping.txt)
personality_mapping = {
    "ESTJ": 0, "ENTJ": 1, "ESFJ": 2, "ENFJ": 3,
    "ISTJ": 4, "ISFJ": 5, "INTJ": 6, "INFJ": 7,
    "ESTP": 8, "ESFP": 9, "ENTP": 10, "ENFP": 11,
    "ISTP": 12, "ISFP": 13, "INTP": 14, "INFP": 15
}

# Apply mapping
df["Personality_Num"] = df["Personality"].map(personality_mapping)

# Check if mapping was successful
df[["Personality", "Personality_Num"]]
```

```python
#Preparing for training
from sklearn.model_selection import train_test_split

#Define features (Excluding Response id and personality_label)
#[Row:Column]
#[Select all rows:,select column excluding 1 and last 2]
X = df.iloc[:,1:-2]
y = df["Personality_Num"] 

#Split dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=50,stratify=y)

#Check data shapes
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from xgboost import XGBClassifier

# Initialize XGBoost classifier with optimal parameters
xgb_model = XGBClassifier(
    objective="multi:softmax",  # Multi-class classification
    num_class=16,  # 16 Personality types
    n_estimators=200,  # Number of trees
    learning_rate=0.05,  # Step size for updating weights
    max_depth=6,  # Tree depth (controls complexity)
    colsample_bytree=0.8,  # Fraction of features used per tree
    subsample=0.8,  # Fraction of training data used per tree
    random_state=42  # Ensures reproducibility
)
```

```python
# Train the model
xgb_model.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Make predictions
y_pred = xgb_model.predict(X_test)

# Measure accuracy
accuracy = accuracy_score(y_test, y_pred)

# Print accuracy
print(f"Model Accuracy: {accuracy:.4f}")

# Detailed classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define hyperparameter grid
param_grid = {
    "n_estimators": [100, 200],  # Number of trees
    "learning_rate": [0.01, 0.05],  # Learning rate
    "max_depth": [4, 6],  # Tree depth
    "subsample": [0.7, 0.8],  # Training sample ratio
    "colsample_bytree": [0.7, 0.8]  # Feature selection ratio
}

# Create the model
xgb_model = XGBClassifier(
    objective="multi:softmax",
    num_class=16,  # Multi-class classification
    eval_metric="mlogloss"  # Metric for optimization
)

# Perform Grid Search
grid_search = GridSearchCV(
    xgb_model, param_grid, cv=3, scoring="accuracy", verbose=2, n_jobs=-1
)

grid_search.fit(X_train, y_train)
```

```python
# Best parameters
print("Best Hyperparameters:", grid_search.best_params_)

# Get the best model
best_xgb = grid_search.best_estimator_

# Make predictions
y_pred_best = best_xgb.predict(X_test)

# Measure accuracy
best_accuracy = accuracy_score(y_test, y_pred_best)

print(f"Optimized Model Accuracy: {best_accuracy:.4f}")
```
